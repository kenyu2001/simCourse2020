```python
import numpy as np
import matplotlib.pyplot as plt

def pi(seed):
    count=0
    
    x=np.random.rand(seed)
    y=np.random.rand(seed)
    color=[]
    
    for i in range(seed):
        xx=1.0-2*x[i]
        yy=1.0-2*y[i]
        #print(xx,yy)
        if xx**2+yy**2<1.0**2:
            count += 1
            color.append('red')
        else:
            color.append('blue')
       
    plt.scatter(1.0-2*x,1.0-2*y,c=color,s=10)
    plt.title('Monte Carlo Simulation for pi')
    
    plt.xlim(xmax=1.0,xmin=-1.0)
    plt.ylim(ymax=1.0,ymin=-1.0)
    #plt.axis("equal")
    plt.gca().set_aspect(1)
    plt.show()
    
    print('pi is ' + str(4*count/seed))

pi(10000)
```
